```python
import ifcopenshell
import re

# # Load IFC file paths from user
ifc = ifcopenshell.open(r"C:\Users\Owner\Documents\25952501_RioSaijo\IFC_ToBrick_IoT_Interoperability_Project\data\In\ifc\curated\51M-SKA-00-00-M3-MEP-0001_IFC 2x3-P01_with_port_relations.ifc")

# Step 1: Find BDNS classification
bdns_classifications = [
    c for c in ifc.by_type("IfcClassification")
    if c.Name == "BDNS"
]

if not bdns_classifications:
    raise Exception("BDNS classification not found")

bdns = bdns_classifications[0]

# Step 2: Get ClassificationReferences belonging to BDNS
bdns_refs = [
    ref for ref in ifc.by_type("IfcClassificationReference")
    if ref.ReferencedSource == bdns
]

bdns_ref_ids = set(r.id() for r in bdns_refs)

# Step 3: Traverse IfcRelAssociatesClassification
for rel in ifc.by_type("IfcRelAssociatesClassification"):
    ref = rel.RelatingClassification

    # Check if the reference belongs to BDNS
    if ref and ref.id() in bdns_ref_ids:
        
# Step 4: Process related instances
        for obj in rel.RelatedObjects:

            if hasattr(obj, "Name") and obj.Name:
                original_name = obj.Name

# Step 5: Remove unnecessary leading zeros
                match = re.match(r"([A-Za-z]+)-0*([0-9]+)", original_name)

                if match:
                    new_name = f"{match.group(1)}-{int(match.group(2))}"

# Step 6: Update the Name attribute
                    obj.Name = new_name

                    print(f"{original_name} -> {new_name}")

# Save modified IFC file
ifc.write(r"C:\Users\Owner\Documents\25952501_RioSaijo\IFC_ToBrick_IoT_Interoperability_Project\data\In\ifc\curated\51M-SKA-00-00-M3-MEP-0001_IFC 2x3-P01_with_port_relations_curated.ifc")
```
